```python
# Histogram of Release Year
plt.figure(figsize=(10, 6))
sb.histplot(df['release_year'], bins=20, kde=True)
plt.title('Distribution of Release Year')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.show()
```

```python
# Countplot of Ratings
plt.figure(figsize=(10, 6))
sb.countplot(data=df, x='rating')
plt.title('Count of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()


```

```python
# Histogram of a numerical column (e.g., 'age')
plt.figure(figsize=(10, 6))
plt.hist(df['duration'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of duration')
plt.xlabel('duration')
plt.ylabel('Frequency')
plt.show()

```

```python
# Data Visualization (example: countplot of type)
plt.figure(figsize=(8, 6))
df['type'].value_counts().plot(kind='bar', color='salmon')
plt.title('Count of Movies vs TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Data Visualization (example: countplot of country)
plt.figure(figsize=(12, 8))
df['rating'].value_counts().head(10).plot(kind='barh', color='b')
plt.title('Top 10 Countries with Most Content')
plt.xlabel('Count')
plt.ylabel('rating')
plt.gca().invert_yaxis()
plt.show()
```

```python
# Data Visualization (example: countplot of country)
plt.figure(figsize=(12, 8))
df['country'].value_counts().head(10).plot(kind='barh', color='lightgreen')
plt.title('Top 10 Countries with Most Content')
plt.xlabel('Count')
plt.ylabel('Country')
plt.gca().invert_yaxis()
plt.show()
```

```python
data=df.groupby(df['country']).sum()
print(data)
```

```python
plt.figure(figsize=(12, 8))
countries = df['country'].value_counts().head(7)  # Get the top 10 countries with most content
plt.barh(countries.index, countries, color='lightgreen')
plt.title('Top 10 Countries with Most Content')
plt.xlabel('Count')
plt.ylabel('Country')
plt.gca().invert_yaxis()
plt.show()
```

```python
countries.index
```

```python
# Data Visualization (example: boxplot of duration)
plt.figure(figsize=(10, 6))
df['duration'] = df['duration'].str.extract('(\d+)').astype(float) # Extract numerical values from duration
plt.boxplot(df['duration'])
plt.title('Boxplot of Duration')
plt.ylabel('Duration (min)')
plt.grid(True)
plt.show()
```

```python
 #Data Visualization (example: pie chart of ratings)
plt.figure(figsize=(15, 9))
df['rating'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightgreen', 'lightcoral', 'lightskyblue'])
plt.title('Distribution of Ratings')
plt.ylabel('')
plt.show()
```

```python
# Drop rows with missing values in the 'cast' column
df.dropna(subset=['cast'], inplace=True)

# Convert 'cast' column to string
df['cast'] = df['cast'].astype(str)

# Data Visualization (word cloud of cast)
from wordcloud import WordCloud

plt.figure(figsize=(10, 8))
wordcloud = WordCloud(background_color='black').generate(' '.join(df['cast']))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Cloud of Cast')
plt.axis('off')
plt.show()
```
